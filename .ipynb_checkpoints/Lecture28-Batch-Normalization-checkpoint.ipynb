## 遇到的问题
1. 损失出现在最后，后面的层训练较快
2. 数据在最底部：
- 底部层的训练比较慢
- 底部层一变化，所有的上面的层都得跟着变
- 最后的那些层需要重新学习多次
- 导致收敛变慢
3. 我们可以在学习底部层的时候避免变换顶部层吗？

## Batch Normalization
先求小批量里的均值和方差，然后再去学习gamma和beta。

因为可能（0，1）的分布未必是最好，所以需要去学习gamma和beta
![softmax-description](./imgs/28-1.png)

## Batch Normalization Layer
- 可学习的参数为gamma和beta
- 作用在：

（1）全连接和卷积层的输出上，激活函数前

原因：其实还是一个线性变化，所以还是在非线性变换之前作用

（2）全连接层和卷积层的输入上

- 对全连接层，作用在特征维（每一行是样本，每一列是特征，对每一个特征计算标量和方差）
- 对卷积层，作用在通道维度（对一个卷积层来说，batch_size* h * w * channel,所有的像素当成是样本，通道维度当作卷积层的特征维，某一通道相当于代表所有样本（batch_size* h * w）的某一特征）

##  Batch Normalization的作用
- 最初论文是想用它去减少内部协变量的转移（后来有人专门去研究，发现并没有减少内部协变量转移）
![softmax-description](./imgs/28-2.png)
- 随机偏移和随机缩放变化剧烈，但是gamma和beta变换还是较小的（基于学习率），因此较稳定
- 因此没必要和drop-out混合使用

##  总结
- Batch Normalization固定小批量中的均值和方差，然后学习出适合的偏移和缩放
-（这样每一层的分布差不多，用统一的学习率，这样每一层的收敛速度也差不多，不会存在底层慢顶层快的情况）
- 可以加速收敛速度，但是一般不改变模型精度

```python
%matplotlib inline
import torch
from torch import nn
import d2l
```

```python
#
def batch_norm(X, gamma, beta, moving_mean, moving_var, eps, momentum):
    # momentum:用于更新moving_mean和moving_var
    # moving_mean, moving_var:整个数据集的均值和方差，用于inference
    # eps:避免除以0
    # 通过 `is_grad_enabled` 来判断当前模式是训练模式还是inference模式
    if not torch.is_grad_enabled():
        # 如果是在预测模式下，直接使用传入的移动平均所得的均值和方差
        # 因为在inference的时候，有时候就传入一个一维样本，可能算不出来均值和方差
        X_hat = (X - moving_mean) / torch.sqrt(moving_var + eps)
    else:
        # 这里是进行训练
        # 2代表全连接层，4代表2维卷积层
        # 我们只考虑这两种情况
        assert len(X.shape) in (2, 4) 
        if len(X.shape) == 2: # batchsize * 特征
            # 使用全连接层的情况，计算特征维上的均值和方差
            # 对每一列求和，dim=0维度保持不变
            mean = X.mean(dim=0)  # mean是一个行向量
            var = ((X - mean) ** 2).mean(dim=0) # var是一个行向量
        else:
            # 使用二维卷积层的情况，计算通道维上（axis=1）的均值和方差。
            # 这里我们需要保持X的形状以便后面可以做广播运算
            mean = X.mean(dim=(0, 2, 3), keepdim=True) # (1*n*1*1)
            var = ((X - mean) ** 2).mean(dim=(0, 2, 3), keepdim=True) ## (1*n*1*1)
        # 训练模式下，用当前的均值和方差做标准化
        X_hat = (X - mean) / torch.sqrt(var + eps)
        # 更新移动平均的均值和方差:在训练的时候会去更新，在inference的时候不会更新
        moving_mean = momentum * moving_mean + (1.0 - momentum) * mean
        moving_var = momentum * moving_var + (1.0 - momentum) * var
    Y = gamma * X_hat + beta  # 缩放和移位
    # Y就是一个线性变换
    # moving_mean，moving_var是torch parameter.
    # 用“.data”返回对应tensor
    return Y, moving_mean.data, moving_var.data
```

```python
class BatchNorm(nn.Module):
    # `num_features`：完全连接层的输出数量或卷积层的输出通道数。
    # `num_dims`：2表示完全连接层，4表示卷积层
    def __init__(self, num_features, num_dims):
        super().__init__()
        # 先求得shape，因为shape会决定gamma,beta,moving_mean,moving_var的shape
        if num_dims == 2:
            shape = (1, num_features)
        else:
            shape = (1, num_features, 1, 1)
        # 参与求梯度和迭代的拉伸和偏移参数，分别初始化成1和0
        # 需要被迭代
        self.gamma = nn.Parameter(torch.ones(shape))
        self.beta = nn.Parameter(torch.zeros(shape))
        # 非模型参数的变量初始化为0和1
        # 不需要被迭代
        self.moving_mean = torch.zeros(shape)
        self.moving_var = torch.ones(shape)

    def forward(self, X):
        # 如果 `X` 不在内存上，将 `moving_mean` 和 `moving_var`
        # 复制到 `X` 所在显存上
        if self.moving_mean.device != X.device:
            self.moving_mean = self.moving_mean.to(X.device)
            self.moving_var = self.moving_var.to(X.device)
        # 保存更新过的 `moving_mean` 和 `moving_var`
        # 不同框架的默认的eps和momentum不同
        Y, self.moving_mean, self.moving_var = batch_norm(
            X, self.gamma, self.beta, self.moving_mean,
            self.moving_var, eps=1e-5, momentum=0.9)
        return Y
```

```python
net = nn.Sequential(
    nn.Conv2d(1, 6, kernel_size=5), 
    BatchNorm(6, num_dims=4), # 加在sigmoid前面
    nn.Sigmoid(),
    nn.MaxPool2d(kernel_size=2, stride=2),
    nn.Conv2d(6, 16, kernel_size=5),
    BatchNorm(16, num_dims=4), 
    nn.Sigmoid(),
    nn.MaxPool2d(kernel_size=2, stride=2), 
    nn.Flatten(),
    nn.Linear(16*4*4, 120), BatchNorm(120, num_dims=2), nn.Sigmoid(),
    nn.Linear(120, 84), BatchNorm(84, num_dims=2), nn.Sigmoid(),
    nn.Linear(84, 10))
```

```python
#这个代码与我们第一次训练 LeNet时几乎完全相同，
# 主要区别在于学习率大得多。
lr, num_epochs, batch_size = 1.0, 10, 256
train_iter, test_iter = d2l.load_data_fashion_mnist(batch_size)
d2l.train_ch6(net, train_iter, test_iter, num_epochs, lr, d2l.try_gpu())
```

```python
# 看看从第一个批量归一化层中学到的拉伸参数 gamma 和偏移参数 beta
net[1].gamma.reshape((-1,)), net[1].beta.reshape((-1,))
```

```python
# 简洁实现
net = nn.Sequential(
    nn.Conv2d(1, 6, kernel_size=5), nn.BatchNorm2d(6), nn.Sigmoid(),
    nn.MaxPool2d(kernel_size=2, stride=2),
    nn.Conv2d(6, 16, kernel_size=5), nn.BatchNorm2d(16), nn.Sigmoid(),
    nn.MaxPool2d(kernel_size=2, stride=2), nn.Flatten(),
    nn.Linear(256, 120), nn.BatchNorm1d(120), nn.Sigmoid(),
    nn.Linear(120, 84), nn.BatchNorm1d(84), nn.Sigmoid(),
    nn.Linear(84, 10))
```

```python
d2l.train_ch6(net, train_iter, test_iter, num_epochs, lr, d2l.try_gpu())
```
